```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

np.random.seed(2014)
A = np.random.randn(5, 5)
G = nx.from_numpy_matrix(A)
list(G.nodes())
```

```python
B = nx.to_numpy_matrix(G)
B
```

```python
plt.matshow(B)
plt.colorbar()
plt.show()
```

```python

A = np.array([[1, 1], [0, 1]])
G = nx.from_numpy_matrix(A)

nx.draw(G, with_labels=True, font_weight='bold')

#show the plot
plt.show()
```
